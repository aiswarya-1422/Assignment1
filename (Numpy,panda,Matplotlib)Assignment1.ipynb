5.Broadcasting
Increase all horsepower values by 10% and store the updated values in a new
NumPy array. Handle missing data (if any) by replacing it with the mean of the
column before applying the increase.

````python
import numpy as np
import pandas as pd

df = pd.read_csv('dataset.csv')


df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
mean_horsepower = np.mean(df['horsepower'])


horsepower = df['horsepower'].fillna(mean_horsepower).to_numpy()

increase_pct = 10
updated_hp = horsepower * (1 + increase_pct / 100)
print(updated_hp)
````

7.Matrix Operations
Create a 2D NumPy array containing the columns mpg, horsepower, and weight.
Compute the dot product of this matrix with a given vector [1, 0.5, -0.2].

````python
import numpy as np
import pandas as pd

df = pd.read_csv('dataset.csv')

# Convert 'horsepower' to numeric, coercing errors to NaN
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Fill missing 'horsepower' values with the mean
mean_horsepower = np.mean(df['horsepower'])
df['horsepower'] = df['horsepower'].fillna(mean_horsepower)

# Create the 2D NumPy array
matrix = np.column_stack((df['mpg'].to_numpy(), df['horsepower'].to_numpy(), df['weight'].to_numpy()))

# Define the vector
vector = np.array([1, 0.5, -0.2])

# Compute the dot product
dot_product = np.dot(matrix, vector)

print(dot_product)
````

8. Sorting
Use NumPy to sort the cars by model_year in descending order and display the first
five car names.

````python
import numpy as np
import pandas as pd

# Load the dataset (assuming df is not already loaded in this cell)
df = pd.read_csv('dataset.csv')

# Get 'model year' and 'car name' as NumPy arrays
model_year_array = df['model year'].to_numpy()
car_name_array = df['car name'].to_numpy()

# Get the indices that would sort 'model year' in descending order
sorted_indices = np.argsort(model_year_array)[::-1]

# Select the car names using the sorted indices and get the first five
sorted_car_names = car_name_array[sorted_indices]
first_five_car_names = sorted_car_names[:5]

# Display the first five car names
print("First five car names sorted by model year (descending):")
print(first_five_car_names)
````

3. Handling Missing Data
Identify if there are any missing values in the dataset. Replace missing values in the
horsepower column with the column's median.

````python
import pandas as pd

df = pd.read_csv('dataset.csv')

# Convert 'horsepower' to numeric, coercing errors to NaN
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

print(df.isnull().sum())

# Calculate the median of the numeric horsepower values
median_horsepower = df['horsepower'].median()

# Fill missing values with the calculated median
df['horsepower'].fillna(median_horsepower, inplace=True)

# Verify that there are no more missing values in 'horsepower'
print(df.isnull().sum())
````

9. Is there a clustering pattern among weight, horsepower, and mpg?

````python
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

df = pd.read_csv('dataset.csv')

df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

mean_horsepower = np.mean(df['horsepower'])
df['horsepower'] = df['horsepower'].fillna(mean_horsepower)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['weight'], df['horsepower'], df['mpg'])
plt.xlabel('Weight')
plt.ylabel('Horsepower')
plt.show()
````
